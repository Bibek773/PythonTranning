```python
import matplotlib.pyplot as plt 
import numpy as np
classes = ["Python", "R" , "AI ", "ML", "DS"]
class1_student = [30,40,50,70,80]
class2_student = [3,48,30,17,58]
class_index = np.arange(len(classes))
width =0.2
plt.bar(class_index, class1_student, width, label = "Class1 Student")
plt.bar(class_index, class2_student, width, label = "Class2 Student")
plt.xticks(class_index+width, classes)
plt.title("Bar CHart of classes")
plt.xlabel("Classes")
plt.ylabel("no of students")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.array([0, 5, 10, 15, 20, 25.35])
plt.plot(x_axis, x_axis * 2)
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.show()
```

```python
plt.plot(x_axis, x_axis**2, 'red' , label= 'x=y**2' )
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
empty_canvas = plt.figure()
y_axis = x_axis**2
create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b')


create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")

create_axis.legend(loc=0)
```
